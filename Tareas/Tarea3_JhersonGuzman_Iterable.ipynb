2. Escriba una clase iterable que tenga la siguiente funcionalidad

    2.1 Al instanciar un objeto, lee el archivo de datos y lo coloca en un dataframe interno en la clase.

    2.2 Cada vez que se le pidan datos, debe regresar los dos siguientes datos.
    
    2.3 Entregue un error si los datos se han agotado

Pruebe su clase.

```python
import pandas as pd

class ClaseIterable:
    def __init__(self):
        self.df = pd.read_csv('/Users/jherguz/Desktop/iris.csv')
        self.index = 0
    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= len(self.df):
            raise StopIteration
        else:
            row = self.df.iloc[self.index]
            self.index += 1
            return row['sepal_length'], row['sepal_width']
```

```python
iris = ClaseIterable()
```

```python
for sepal_length, sepal_width in iris:
    print(sepal_length, sepal_width)
```
